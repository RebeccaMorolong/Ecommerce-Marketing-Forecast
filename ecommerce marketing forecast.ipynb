```python
#creage a database named website_data
try:
    conn = psycopg2.connect(
        dbname="postgres",
        user="postgres",
        password="Shane223@@",
        host="localhost",
        port=5432
    )
    conn.autocommit = True  # Enable autocommit mode

    cur = conn.cursor()
    cur.execute("CREATE DATABASE website_data;")
except psycopg2.Error as e:
    print(f"Error creating database: {e}")

# Close the connection to the initial database
cur.close()
conn.close()
# Import the psycopg2 library for PostgreSQL database interaction
import psycopg2
# Ensure the psycopg2 library is installed
# If not installed, you can install it using pip:
# pip install psycopg2-binary
# Import the psycopg2 library for PostgreSQL database interaction
import psycopg2
# Ensure the psycopg2 library is installed
# If not installed, you can install it using pip:
# pip install psycopg2-binary
# Connect to the PostgreSQL database and import dataset into a table named website_data
conn = psycopg2.connect(
    dbname="website_data",
    user="postgres",
    password="Shane223@@",
    host="localhost",
    port=5432
)
cur = conn.cursor()
# Create a table if it doesn't exist
cur.execute("""
CREATE TABLE IF NOT EXISTS website_data (
    id SERIAL PRIMARY KEY,
    customer_behavior FLOAT,
    market_trends FLOAT,
    seasonal_fluctuations TEXT,
    product_availability INTEGER,
    customer_demographics TEXT,
    website_traffic INTEGER,
    engagement_rate FLOAT,
    sales_forecast INTEGER
);
""")

# Insert data into the table
import csv

data_file = r'c:\Users\Rebecca\OneDrive\Desktop\MALESHANE\MALESHANE\ecommerce_sales_forecast_dataset.csv'

with open(data_file, newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        cur.execute("""
        INSERT INTO website_data (
            customer_behavior,
            market_trends,
            seasonal_fluctuations,
            product_availability,
            customer_demographics,
            website_traffic,
            engagement_rate,
            sales_forecast
        ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s);
        """, (
            float(row['Customer Behavior']),
            float(row['Market Trends']),
            row['Seasonal Fluctuations'],
            int(row['Product Availability']),
            row['Customer Demographics'],
            int(row['Website Traffic']),
            float(row['Engagement Rate']),
            int(row['Sales Forecast'])
        ))

# Commit the transaction
conn.commit()
```
